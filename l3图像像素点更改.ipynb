```python
gege = cv.imread("gege.jpg")
```

```python
plt.imshow(gege[:,:,::-1])
```

```python
b,g,r = cv.split(gege)
```

```python
plt.imshow(b,cmap=plt.cm.gray)
```

```python
img2 = cv.merge((b,g,r))
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
gray = cv.cvtColor(gege,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap=plt.cm.gray)
```

```python
hsv = cv.cvtColor(gege,cv.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv)
```
