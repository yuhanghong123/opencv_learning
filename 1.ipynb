```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread("test01.jpg")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
ret ,thresh = cv.threshold(gray,120,255,cv.THRESH_BINARY)

plt.imshow(thresh,cmap=plt.cm.gray)
```

```python
contours,hierarchy =cv.findContours(thresh,cv.RETR_TREE,cv.CHAIN_APPROX_NONE)
draw_img = img.copy()
res = cv.drawContours(draw_img,contours,4,(255,0,0),1)
plt.figure(figsize=(10,10))
plt.imshow(res[:,:,::-1])
```

```python
coners = cv.goodFeaturesToTrack(gray,4,0.02,2)
```

```python
coners = np.intp(coners)
```

```python
for i in coners:
    x,y= i.ravel()
    cv.circle(img,(x,y),2,(0,0,255),-1)
```
