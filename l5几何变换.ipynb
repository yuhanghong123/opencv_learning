```python
import matplotlib.pyplot as plt 
import numpy as np 
import cv2 as cv 
```

```python
gege = cv.imread("gege.jpg")
```

```python
plt.imshow(gege[:,:,::-1])
```

```python
# 
rows,cols = gege.shape[:2]
```

```python
rows
```

```python
cols
```

```python
res = cv.resize(gege,(2*cols,2*rows))
```

```python
plt.imshow(res[:,:,::-1])
```

```python
# 
res1 = cv.resize(gege,None,fx=0.5,fy=0.5)
```

```python
plt.imshow(res1[:,:,::-1])
```

```python
res1.shape
```

# 

```python
M = np.float32([[1,0,100],[0,1,50]])
```

```python
res2 = cv.warpAffine(gege,M,(2*cols,2*rows))
```

```python
plt.imshow(res2[:,:,::-1])
```

# 

```python
M = cv.getRotationMatrix2D((cols/2,rows/2),45,1)
```

```python
res3 = cv.warpAffine(gege,M,(cols,rows))
```

```python
plt.imshow(res3[:,:,::-1])
```

# 

```python
pts1 = np.float32([[50,50],[200,50],[250,70]])
```

```python
pts2 = np.float32([[100,100],[200,50],[100,50]])
```

```python
M = cv.getAffineTransform(pts1,pts2)
```

```python
res4 = cv.warpAffine(gege,M,(cols,rows))
```

```python
plt.imshow(res4[:,:,::-1])
```

# 

```python
pts1 = np.float32([[55,23],[32,43],[100,50],[200,100]])
```

```python
pts2 = np.float32([[64,32],[44,66],[120,70],[220,120]])
```

```python
T = cv.getPerspectiveTransform(pts1,pts2)
```

```python
T
```

```python
res5 = cv.warpPerspective(gege,T,(cols,rows))
```

```python
plt.imshow(res5[:,:,::-1])
```
