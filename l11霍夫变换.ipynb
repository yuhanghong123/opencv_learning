```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv 
```

# 1 线检测

```python
img = cv.imread("date.jpg")
```

```python
edges = cv.Canny(img,50,150)
```

```python
plt.imshow(edges,cmap=plt.cm.gray)
```

```python
lines = cv.HoughLines(edges,0.8,np.pi/180,150)
```

```python
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = rho*a
    y0 = rho*b
    x1 = int(x0+1000*(-b))
    y1 = int(y0+1000*(a))
    x2 = int(x0-1000*(-b))
    y2 = int(y0-1000*a)
    cv.line(img,(x1,y1),(x2,y2),(0,255,0))
```

```python
plt.imshow(img[:,:,::-1])
```

# 霍夫圆检测

```python
planet = cv.imread("planet.jpg")
```

```python
gray_img = cv.cvtColor(planet,cv.COLOR_BGR2GRAY)
```

```python
img = cv.medianBlur(gray_img,7)
```

```python
plt.imshow(img,cmap=plt.cm.gray)
```

```python
circles = cv.HoughCircles(img,cv.HOUGH_GRADIENT,1,200,param1=100,param2=50,minRadius=0,maxRadius=100)
```

```python
circles
```

```python
for i in circles[0,:]:
    cv.circle(planet,(i[0],i[1]),i[2],(0,255,0),2)
    # cv.circle(planet,(i[0],i[1]),2,(0,255,0),-1)
```
