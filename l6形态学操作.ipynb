```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread("i.jpg")
```

```python
plt.imshow(img[:,:,::-1])
```

# 

```python
kernel = np.ones((5,5),np.uint8)
```

```python
img2 = cv.erode(img,kernel)
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
# 
img1 = cv.dilate(img,kernel)
```

```python
plt.imshow(img1[:,:,::-1])
```

# 

```python
open = cv.imread("kai.jpg")
```

```python
plt.imshow(open[:,:,::-1])
```

```python
close = cv.imread("bi.jpg")
```

```python
kernal = np.ones((10,10),np.uint8)
```

```python
cvopen = cv.morphologyEx(open,cv.MORPH_OPEN,kernal)
```

```python
plt.imshow(cvopen[:,:,::-1])
```

```python
plt.imshow(close[:,:,::-1])
```

```python
cvclose = cv.morphologyEx(close,cv.MORPH_CLOSE,kernal)
```

```python
plt.imshow(cvclose[:,:,::-1])
```

# 

```python
top = cv.morphologyEx(open,cv.MORPH_TOPHAT,kernal)
```

```python
plt.imshow(top[:,:,::-1])
```

```python
black = cv.morphologyEx(close,cv.MORPH_BLACKHAT,kernal)
```

```python
plt.imshow(black[:,:,::-1])
```
