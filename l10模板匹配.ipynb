```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv 
```

```python
img1 = cv.imread("wulin.jpg")
```

```python
img2 = cv.imread("moban.jpg")
```

```python
plt.imshow(img1[:,:,::-1])
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
res = cv.matchTemplate(img1,img2,cv.TM_CCORR)
```

```python
plt.imshow(res,cmap=plt.cm.gray)
```

```python
min_val,max_val,min_loc,max_loc = cv.minMaxLoc(res)
```

```python
top_left = max_loc
```

```python
h,w = img2.shape[:2]
```

```python
bottom_right = (top_left[0]+w,top_left[1]+h)
```

```python
cv.rectangle(img1,top_left,bottom_right,(0,255,0),2)
```

```python
plt.imshow(img1[:,:,::-1])
```
