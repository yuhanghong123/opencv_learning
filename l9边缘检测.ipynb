```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt 
```

```python
house = cv.imread("house.jpg",0)
```

```python
plt.imshow(house,cmap=plt.cm.gray)
```

 # Sobel

```python
x = cv.Sobel(house,cv.CV_16S,1,0)
y = cv.Sobel(house,cv.CV_16S,0,1)
```

```python
abx = cv.convertScaleAbs(x)
aby = cv.convertScaleAbs(y)
```

```python
res = cv.addWeighted(abx,0.5,aby,0.5,0)
```

```python
plt.imshow(res,cmap=plt.cm.gray)
```

# Schaar

```python
x = cv.Sobel(house,cv.CV_16S,1,0,ksize=-1)
y = cv.Sobel(house,cv.CV_16S,0,1,ksize=-1)
```

```python
abx = cv.convertScaleAbs(x)
aby = cv.convertScaleAbs(y)
```

```python
res = cv.addWeighted(abx,0.5,aby,0.5,0)
```

```python
plt.imshow(res,cmap=plt.cm.gray)
```
