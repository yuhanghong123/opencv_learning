```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
img = cv.imread("chess.jpg")
```

```python
plt.imshow(img[:,:,::-1])
```

# Harris

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
gray = np.float32(gray)
```

```python
dst = cv.cornerHarris(gray,2,3,0.04)
```

```python
img[dst>0.001*dst.max()]=[255,0,0]
```

```python
plt.imshow(img[:,:,::-1])
```

# shi-Tomasi

```python
img = cv.imread("tv.jpg")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
coners = cv.goodFeaturesToTrack(gray,300,0.01,10)
```

```python
coners = np.int0(coners)
```

```python
for i in coners:
    x,y = i.ravel()
    cv.circle(img,(x,y),2,(255,0,0),-1)
```

```python
plt.imshow(img[:,:,::-1])
```

# SIFT

```python
img = cv.imread("tv.jpg")
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
sift = cv.SIFT_create()
```

```python
kp,des = sift.detectAndCompute(gray,None)
```

```python
cv.drawKeypoints(img,kp,img,flags=cv.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
```

```python
plt.imshow(img[:,:,::-1])
```

```python
img = cv.imread("tv.jpg")
```

```python
fast = cv.FastFeatureDetector_create(threshold=30)
```

```python
kp =fast.detect(img,None)
```

```python
img2 = cv.drawKeypoints(img,kp,None,color=(255,0,0))
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
fast.setNonmaxSuppression(0)
```

```python
kp = fast.detect(img,None)
```

```python
img3 = cv.drawKeypoints(img,kp,None,color=(255,0,0))
```

```python
plt.imshow(img3[:,:,::-1])
```
