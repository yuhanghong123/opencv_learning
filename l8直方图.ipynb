```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2 as cv
```

```python
cat = cv.imread("cat.jpg",0)
```

```python
plt.imshow(cat,cmap=plt.cm.gray)
```

# 

```python
hist = cv.calcHist([cat],[0],None,[256],[0,256])
```

```python
plt.figure(figsize=(10,8))
plt.plot(hist)
plt.show()
```

```python
mask = np.zeros(cat.shape[:2],np.uint8)
mask[160:240,80:180] = 1
```

```python
plt.imshow(mask,cmap= plt.cm.gray)
```

```python
mask_img = cv.bitwise_and(cat,cat,mask=mask)
```

```python
plt.imshow(mask_img,cmap=plt.cm.gray)
```

```python
mask_hist = cv.calcHist([cat],[0],mask,[256],[0,256])
```

```python
plt.plot(mask_hist)
plt.show()
```

# 

```python
dst = cv.equalizeHist(cat)
```

```python
plt.imshow(dst,cmap=plt.cm.gray)
```

```python
cl = cv.createCLAHE(2.0,(8,8))
```

```python
clahe = cl.apply(cat)
```

```python
plt.imshow(clahe,cmap=plt.cm.gray)
```
